```python
import numpy as np
import pylab as plt
from scipy.optimize import curve_fit

def f(x, a, b, c):
    return a * np.exp(-b * x) + c

#generowanie danych "idealnych"
x = np.linspace(0, 4, 50)
y = f(x, 2.5, 1.3, 0.5)

#dodanie szumu
yi = y + 0.2 * np.random.normal(size=len(x))

#Wywołanie funkcji curve_fit
popt, pcov = curve_fit(f, x, yi)
a, b, c = popt
#Wypisanie wyników
print(" Parametry optymalne: ", popt)

yfitted = f(x, *popt) # zapis rozwnowazny z f(x, popt[0], popt[1], popt[2])
plt.plot(x, yi, 'o', label='dane $y_i$')
plt.plot(x, yfitted, '-', label='dopasowanie $f(x_i)$')
plt.xlabel
plt.legend()
plt.show()

"""

funkcja curve_fit() jest świetnym narzędziem do dopasowywania modelu matematycznego do funkcji.
Pozwala ona na określenie optymalnych wartości parametrów modelu na podstawie danych eksperymentalnych.
Dodawanie szumu losowego do danych pomiarowych pozwala na symulację rzeczywistych warunków pomiarowych uwzględniając naturalną zmiennośc i błędy pomiarowe.
Dopasowanie modelu do danych pozwala na określenie parametrów modelu, które najlepiej opisują dane.

"""
```
